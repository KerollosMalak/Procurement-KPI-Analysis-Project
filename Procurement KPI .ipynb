```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('E:/data analysis matirals/TEST/data/Procurement KPI Analysis Dataset.csv')
df
```

```python
df.dropna(subset=['Delivery_Date'], inplace=True)
df
```

```python
df['Delivery_Date']=pd.to_datetime(df['Delivery_Date'],format='mixed')
df['Order_Date']=pd.to_datetime(df['Order_Date'],format='mixed')
df
```

```python
df['duration'] = (df['Delivery_Date'] - df['Order_Date']).dt.days
df
```

```python
df['%_of_Defective_Units']=((df['Defective_Units']/df['Quantity'])*100)
```

```python
df
```

```python
df['%_of_Defective_Units'] = df['%_of_Defective_Units'].fillna(0)
df['Defective_Units'] = df['Defective_Units'].fillna(0)
df
```

```python
df.isna().sum()
```

```python
df['Total_Spend'] =( df['Quantity'] * df['Unit_Price'])
df
```

```python
top_suppliers = df.groupby('Supplier')['Total_Spend'].sum().sort_values(ascending=False)
df
```

```python
df['Order_Month'] = df['Order_Date'].dt.to_period('M')
monthly_trends = df.groupby('Order_Month')['Total_Spend'].sum()
monthly_trends.plot(kind='bar', title='Monthly Procurement Spend')
```

```python
avg_duration = df['duration'].median()
print(f'Average Delivery Duration: {avg_duration} days')
```

```python
late_deliveries = df[df['duration'] > avg_duration]
print(f'Number of late deliveries: {late_deliveries.shape[0]}')
```

```python
df['Order_Month'] = df['Order_Date'].dt.to_period('M')

monthly_orders = df.groupby('Order_Month')['PO_ID'].count()
monthly_orders.plot(kind='line', title='Monthly Purchase Orders')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_suppliers.values, y=top_suppliers.index, palette="Blues_r")
plt.title("Top 10 Suppliers by Total Spend")
plt.xlabel("Total Spend")
plt.ylabel("Supplier")
plt.tight_layout()
plt.show()
```

```python
top_defect_rates = df.groupby('Supplier')['%_of_Defective_Units'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_defect_rates.values, y=top_defect_rates.index, palette="Blues_r")
plt.title("Top 10 Suppliers by Average Defective Rate (%)")
plt.xlabel("Average % Defective Units")
plt.ylabel("Supplier")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
monthly_orders.plot(kind='line', marker='o', color='Blue')
plt.title("Monthly Purchase Order Volume")
plt.xlabel("Month")
plt.ylabel("Number of POs")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=category_qty.values, y=category_qty.index, palette="Greens_r")
plt.title("Top 10 Most Ordered Item Categories")
plt.xlabel("Quantity Ordered")
plt.ylabel("Item Category")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
compliance_counts = df['Compliance'].value_counts()
compliance_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.title("Compliance Distribution")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
df['Discount'] = df['Unit_Price'] - df['Negotiated_Price']
discount_avg = df.groupby('Supplier')['Discount'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=discount_avg.values, y=discount_avg.index, palette="Purples_r")
plt.title("Top 10 Suppliers by Average Discount Achieved")
plt.xlabel("Average Discount (Unit Price - Negotiated Price)")
plt.ylabel("Supplier")
plt.tight_layout()
plt.show()
```

```python
supplier_volume = df['Supplier'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=supplier_volume.values, y=supplier_volume.index, palette="Oranges_r")
plt.title("Top 10 Suppliers by Number of Purchase Orders")
plt.xlabel("Number of Orders")
plt.ylabel("Supplier")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Order_Status', y='duration', data=df, palette='Set2')
plt.title("Delivery Duration by Order Status")
plt.xlabel("Order Status")
plt.ylabel("Delivery Duration (days)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['Quantity', 'Unit_Price', 'Negotiated_Price', 'Defective_Units', 'duration', '%_of_Defective_Units']].corr(), 
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.show()
```
